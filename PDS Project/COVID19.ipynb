# Imports

```python
import matplotlib.pyplot as plt
from matplotlib import style # For some styling
import pandas as pd
```

# Reading the data

```python
data = pd.read_csv('covid_day_wise.csv')
```

```python
data
```

```python
data.columns
```

```python
data.info()
```

# Data preparation
## Check null values

```python
data.isna().sum()
```

No Null values in the data(Thank god)

## Convert Date to datetime data type

```python
# Before conversion
data.info()
```

```python
# Uses the .to_datetime() function to convert 
data["Date"] = pd.to_datetime(data['Date'])
```

```python
# After conversion
data.info()
```

## Set Date as index

```python
data = data.set_index("Date")
```

## Sort the new index

```python
data.sort_index(ascending=True, inplace=True)
data
```

# Visualization
## Make a line plot for every column available in the data (Date vs. column)

```python
# Number of countries 
data.plot.line(y="No. of countries")
plt.xlabel("Date", size=20)
plt.ylabel("Infected countries", size=20)
plt.title("Number of infected countries", size=25)
plt.style.use('ggplot')
```

```python
# Confirmed cases plot 
data.plot.line(y="Confirmed")
plt.xlabel("Date", size=20)
plt.ylabel("Cases", size=20)
plt.title("Number of total confirmed cases", size=25)
plt.style.use('ggplot')
```

```python
# Deaths plot 
data.plot.line(y="Deaths")
plt.xlabel("Date", size=20)
plt.ylabel("Deaths", size=20)
plt.title("Number of total deaths", size=25)
plt.style.use('ggplot')
```

```python
# Recovered plot 
data.plot.line(y="Recovered")
plt.xlabel("Date", size=20)
plt.ylabel("Recoveries", size=20)
plt.title("Number of total recoveries", size=25)
plt.style.use('ggplot')
```

```python
# Active plot 
data.plot.line(y="Active")
plt.xlabel("Date", size=20)
plt.ylabel("Active Cases", size=20)
plt.title("Number of total active cases", size=25)
plt.style.use('ggplot')
```

```python
# new cases plot 
data.plot.line(y="New cases")
plt.xlabel("Date", size=20)
plt.ylabel("New Cases", size=20)
plt.title("Number of new cases per day", size=25)
plt.style.use('ggplot')
```

```python
# new death plot 
data.plot.line(y="New deaths")
plt.xlabel("Date", size=20)
plt.ylabel("New Deaths", size=20)
plt.title("Number of new deaths per day", size=25)
plt.style.use('ggplot')
```

```python
# new recovered plot 
data.plot.line(y="New recovered")
plt.xlabel("Date", size=20)
plt.ylabel("New Recoveries", size=20)
plt.title("Number of new recoveries per day", size=25)
plt.style.use('ggplot')
```

## Plot a line plot: Date vs. (Deaths/Confirmed)
<p>Deaths/Confirmed is the total fatality rate</p>

```python
# Fatality rate plot
fatality_rate = data["Deaths"] / data["Confirmed"]
fatality_rate.plot.line()
plt.xlabel("Date", size=20)
plt.ylabel("Rate", size=20)
plt.title("Fatality rate", size=25)
plt.style.use('ggplot')
```

## Plot a line plot: Date vs. (Recovered/Confirmed)
<p>Revocered/Confirmed is the total survival rate</p>

```python
survival_rate = data["Recovered"] / data["Confirmed"]
survival_rate.plot.line()
plt.xlabel("Date", size=20)
plt.ylabel("Rate", size=20)
plt.title("Survival rate", size=25)
plt.style.use('ggplot')
```
