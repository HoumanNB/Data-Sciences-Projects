```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating random data with normal distribution
x = np.random.randn(10000)
```

```python
# Crating the array for plot with 100 edges
hist, bin_edges = np.histogram(x, bins = 100)
```

```python
# Creating the plot
fig, ax = plt.subplots()

ax.hist(x, bin_edges, cumulative = False)

ax.set_xlabel('X')
ax.set_ylabel('Frequency')
```
